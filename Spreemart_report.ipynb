### Model Features Summary

| Feature                          | Description                                                 |

| `payment_type_online`            | Indicator of online payment method usage.                   |
| `payment_type_voucher`           | Indicator of voucher usage for payments.                    |

## Customer Engagement Activities by Segment


##### Occasional Buyers
- **Loyalty Programs**: Implement a loyalty program that rewards frequent purchases, even if they are of lower value, to encourage repeat business.
- **Special Offers**: Provide exclusive deals or discounts on items they frequently purchase.
- **Email Marketing**: Use email marketing campaigns to inform them of new products or sales in categories they have shown interest in.
- **Personalized Recommendations**: Based on their purchase history, offer personalized product recommendations to encourage exploration and repeat visits.

##### Strategic Shoppers
- **Re-engagement Campaigns**: Since these customers are infrequent, use re-engagement tactics like special "we miss you" offers or reminders of items they've viewed but not purchased.
- **Seasonal Promotions**: Target them with promotions around specific holidays or seasons when they are more likely to make purchases.
- **Feedback Solicitation**: Request feedback on their purchases to build engagement and gather insights for future offerings.

##### Loyal Customers
- **Exclusive Access**: Give them early access to new products or sales as a reward for their loyalty.
- **Customer Community**: Invite them to join a customer community or forum where they can discuss products and offer suggestions, enhancing their sense of belonging.
- **Referral Programs**: Encourage them to refer friends or family, rewarding them for bringing new customers to your business.

##### Premium Customers
- **Premium Customer Experience**: Offer a premium shopping experience with personalized services like a dedicated account manager or personalized shopping assistance.
- **High-Value Product Updates**: Notify them of new arrivals or exclusive items in the high-value product range.
- **Membership Programs**: Create a VIP membership program that offers exclusive benefits, such as faster shipping, extended return periods, or exclusive events.

### Customer Segment Distribution

Show the distribution of customers across different segments.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# read customer_segmented.csv
customer_segmented = pd.read_csv('./cache/customer_segmented.csv')

# Mapping of segment IDs to names
segment_names = {
    0: 'Occasional Buyers',
    1: 'Strategic Shoppers',
    2: 'Loyal Customers',
    3: 'Premium Customers'
}

customer_segmented['Segment'] = customer_segmented['Segment'].map(segment_names)

# print(customer_segmented.head())
```

```python
segment_distribution = customer_segmented['Segment'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(segment_distribution, labels=segment_distribution.index, autopct='%1.1f%%')
plt.title('Customer Segment Distribution')
plt.show()
```


The RentSpree Customer Segmentation Report shows a diverse mix of customer groups. "Occasional Buyers" and "Loyal Customers" make up significant equal portions of our clientele, while "Strategic Shoppers" and "Premium Customers" represent smaller, yet important segments. This variety highlights the need for tailored strategies to effectively engage each group.

### Average Spend by Segment

Show how different segments contribute to the overall spending.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Segment', y='total_spend', data=customer_segmented)
plt.title('Average Total Spend by Segment')
plt.ylabel('Average Total Spend')
plt.xlabel('Segment')
plt.show()

```

Premium Customers are the highest spenders, with an average spend more that $400. Strategic Shoppers,  Loyal Customers and Occasional Buyers spend around $200 on average.

### Payment Method Preferences

Show the preferred payment methods of each segment.

```python
payment_columns = ['payment_type_credit_card', 'payment_type_debit_card', 'payment_type_online', 'payment_type_voucher']
payment_preferences = customer_segmented.groupby('Segment')[payment_columns].mean()

payment_preferences.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Payment Method Preferences by Segment')
plt.ylabel('Average Count')
plt.xlabel('Segment')
plt.show()

# print(payment_preferences)
```

"Loyal Customers" and "Strategic Shoppers" mostly use credit cards, with some interest in vouchers, especially among Strategic Shoppers. "Occasional Buyers" are similar but use online payments a bit more. "Premium Customers" stand out for mainly using credit cards.

### Customer Churning Analysis

We define potential churn as customers who haven't made a purchase in 180 days.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define a threshold for potential churn
churn_threshold = 180

customer_segmented['Potential_Churn'] = customer_segmented['time_since_last_purchase'] >= churn_threshold
churn_rate_by_segment = customer_segmented.groupby('Segment')['Potential_Churn'].mean()
```

```python
# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=churn_rate_by_segment.index, y=churn_rate_by_segment.values)
plt.title('Potential Customer Churn Rate by Segment')
plt.ylabel('Churn Rate')
plt.xlabel('Segment')
plt.xticks(rotation=45)  # Rotate segment names for readability
plt.show()

# print(churn_rate_by_segment)
```

- **Loyal Customers**: shows a 0% churn rate, indicating strong customer retention and consistent engagement.
- **Occasional Buyers**: high potential churn rate at 93.6%. It suggests that these customers engage infrequently, possibly only for specific, sporadic needs.
- **Premium Customers**: churn rate of 48.4%, there's a considerable turnover in this segment. While these customers may be high-value, their loyalty appears less consistent.
- **Strategic Shoppers**: shows a 100% potential churn rate, which could indicate a critical engagement needed.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Reshape time of day 
time_of_day = customer_segmented.melt(id_vars=['Segment'], 
                                      value_vars=['purchase_time_of_day_Afternoon', 
                                                  'purchase_time_of_day_Evening', 
                                                  'purchase_time_of_day_Morning', 
                                                  'purchase_time_of_day_Night'],
                                      var_name='Time_of_Day', 
                                      value_name='Count')

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=time_of_day, x='Time_of_Day', y='Count', hue='Segment')
plt.title('Purchase Time of Day by Customer Segment')
plt.ylabel('Average Count')
plt.xlabel('Time of Day')
plt.xticks(rotation=45)
plt.legend(title='Segment')
plt.show()
```

All customer segments tend towards making purchases in the afternoon and evening, with the afternoon being slightly more popular. All segments show decreased activity in the morning and even less at night. This pattern suggests that marketing efforts and customer service resources may be most effectively allocated during the afternoon and evening hours when customers are most active. The lower activity in the mornings and nights might indicate an opportunity to capture a market segment that prefers non-peak hours for shopping, potentially offering incentives or special deals during these times to increase sales.

Note:
- Morning: 6am - 12pm
- Afternoon: 12pm - 6pm
- Evening: 6pm - 12am
- Night: 12am - 6am

```python
# Reshape day of week analysis
day_of_week = customer_segmented.melt(id_vars=['Segment'], 
                                      value_vars=['purchase_day_of_week_Weekday', 
                                                  'purchase_day_of_week_Weekend'],
                                      var_name='Day_of_Week', 
                                      value_name='Count')

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=day_of_week, x='Day_of_Week', y='Count', hue='Segment')
plt.title('Purchase Day of Week by Customer Segment')
plt.ylabel('Average Count')
plt.xlabel('Day of Week')
plt.xticks(rotation=45)
plt.legend(title='Segment')
plt.show()
```

The purchases are more frequently made on weekdays compared to weekends. This suggests that customers prefer to shop on days when they may also have work-related tasks, possibly during lunch breaks or right before or after work hours. The relatively lower activity on weekends could indicate that customers spend these days away from online shopping or use them for other leisure activities. The consistency across segments suggests that this is a broader behavioral pattern and not restricted to specific types of customers. 

This insight could inform decisions on when to schedule marketing campaigns, with a focus on weekday promotions to capitalize on higher engagement rates. Additionally, it may influence operational planning, such as customer support and logistics, ensuring that resources are properly allocated during higher traffic periods.
